```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Employee_Salary_Dataset.csv")
```

```python
df.columns=df.columns.str.strip()
df.columns=df.columns.str.replace('_','')
```

```python
print(df.head())
print(df.info())
print(df.describe())
```

```python
df.isnull().sum()
df=df.dropna()
```

```python
plt.figure()
df.hist(figsize=(10,8))
plt.show()
```

```python
print("Average:",df['Salary'].mean())
print("Median:",df['Salary'].median())
print("Max:",df['Salary'].max())
print("Min:",df['Salary'].min())
```

# Note: Dataset does not contain department or role column
# so comparison cannot be performes

```python
print(df.sort_values(by='Salary',ascending=False).head())
```

```python
plt.figure()
plt.hist(df['Salary'])
plt.title("Salary Distribution")
plt.show()
```

```python
plt.figure()
plt.scatter(df['ExperienceYears'],df['Salary'])
plt.xlabel("Experince")
plt.ylabel("Salary")
plt.title("Experience vs Salary")
plt.show
```

```python
sns.boxplot(x=df['Salary'])
```

```python
# Insights
# 1. Salary increases with Experience
# 2. Some outliers have very high Salaries
# 3. Majority Employees fall in mid Salary range
```
